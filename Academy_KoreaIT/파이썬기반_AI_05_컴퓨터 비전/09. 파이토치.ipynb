### 1-1. Tensor
* 텐서는 배열이나 행렬과 매우 유사한 특수 자료구조
* 파이토치는 텐서를 사용하여 모델의 입력과 출력, 모델의 매개변수들을 처리

```python
data = [[1, 2], [3, 4]]
x_data = torch.tensor(data)
print(x_data)
```

```python
import numpy as np
```

```python
np_array = np.array(data)
x_np_1 = torch.tensor(np_array)
print(x_np_1)
```

```python
x_np_1[0, 0] = 100
print(x_np_1)   # 새로운 텐서
print(np_array) # 기존 ndarray
```

```python
x_np_2 = torch.as_tensor(np_array)  # ndarray와 동일한 메모리 주소를 가리키는 뷰를 만듦
print(x_np_2)
x_np_2[0, 0] = 200  # 기존 메모리 주소의 ndarray값을 변경하게됌
print(x_np_2)
print(np_array)
```

```python
x_np_3 = torch.from_numpy(np_array)  # ndarray와 동일한 메모리 주소를 가리키는 뷰를 만듦
print(x_np_3)
x_np_3[0, 0] = 400  # 기존 메모리 주소의 ndarray값을 변경하게됌
print(x_np_3)
print(np_array)
```

```python
np_again = x_np_1.numpy()
print(np_again, type(np_again))
```

### 1-2. Tensor 속성

```python
tensor = torch.rand(3, 4)
print(f'shape : {tensor.shape}')
print(f'type : {tensor.dtype}')
print(f'device : {tensor.device}')
```

### 1-3. GPU 사용하기
* 코랩에서 device 변경하는 방법
    * 상단 메뉴 → 런타임 → 런타임 유형변경 → 하드웨어 가속기를 GPU로 변경 → 저장 → 다시 시작 및 모두 실행

```python
tensor = tensor.reshape(4, 3)
tensor = tensor.int()
if torch.cuda.is_available():
    print('GPU를 사용할 수 있음')
    tensor = tensor.to('cuda')

print(f'shape : {tensor.shape}')
print(f'type : {tensor.dtype}')
print(f'device : {tensor.device}')
```

### 1-4. 인덱싱과 슬라이싱

```python
a = torch.arange(1, 13).reshape(3, 4)
print(a)
```

```python
print(a[1])
print(a[0, -1])
print(a[1:-1])
print(a[:2, 2:])
```

### 1-5. Transpose
* 두 개의 차원을 맞교환할 때 사용

```python
a = torch.arange(16).reshape(2, 2, 4)
print(a, a.shape)
```

```python
b = a.transpose(1, 2)
print(b, b.shape)
```

### 1-6. permute

```python
# a → 2, 2, 4
c = a.permute((2, 0, 1)) # 4, 2, 2
print(c, c.shape)
```

### 1-7. Tensor 연산

```python
x = torch.tensor([[1, 2], [3, 4]], dtype=torch.float32)
y = torch.tensor([[1, 2], [3, 4]], dtype=torch.float32)
print(x)
print(y)
```

```python
print(torch.add(x, y))
print(torch.subtract(x, y))
print(torch.multiply(x, y))
print(torch.divide(x, y))
print(torch.matmul(x, y))
```

```python
print(x + y)
print(x - y)
print(x * y)
print(x / y)
print(x @ y)
```

```python
# in-place 연산
# ex) num += 1
print(x.add(y))
print(x)
print(y)
```

```python
print(x.add_(y)) # x에 결과를 다시 저장
print(x)
# 위 모든 연산자에 _를 붙여 사용할 수 있음
```

```python
z = torch.arange(1, 11).reshape(2, 5)

sum1 = torch.sum(z, axis=0)
print(sum1, sum1.shape)
```
